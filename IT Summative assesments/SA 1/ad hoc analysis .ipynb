```python

#january



januarydfrowidx=items_for_cleanup['transaction_date'].str.contains('2022/01')

januarydfcolidx=['transaction_date','transaction_items','transaction_value']

januarydf=items_for_cleanup.loc[januarydfrowidx,januarydfcolidx]


# def commafix(transaction):
#     transaction_items=transaction.replace(',(',' (')
#     transaction_items=transaction_items.replace(';',',')
   
#     return(transaction_items)

# januarydf['transaction_items']=np.vectorize(commafix)(januarydf['transaction_items'])

# itemlist=fixedstring.split(',')
# def counter(fixedstring):
januarylist=[]
for transaction in zip(januarydf['transaction_items']): 
    

    trial_list=transaction[0].replace(',(',' (')

    trial_list=trial_list.replace(';',',')

    trial_list=trial_list.split(',')
    for item in trial_list:
        count=1
        paren_variable=-1
        item_value=item
        if item.find('(')!=(-1):
            paren_variable=item.index('(')
            index=paren_variable+2
            count=int(item[index:-1])
            item_value=(item[0:paren_variable-1])
        january_list.append({'item':item_value,'count':count})
        
january_df_final=pd.DataFrame(january_list)

january_df_final
    
january_df_final['Total items']=january_df_final.groupby(['item'])['count'].transform('sum')
    
    
january_total_items_df_raw=january_df_final.drop_duplicates(subset=['item', 'Total items'])
        
january_total_items_df=january_total_items_df_raw.copy()



finaljanuary_total_items_df=january_total_items_df.drop(columns=['count'])


print('TOTAL ITEMS SOLD FOR JANUARY')

finalfinaljan=finaljanuary_total_items_df.drop(df.index[[0,4,6]])

#exotic extras is a brand
#candy city is a brand
#healthykid3+


#multiplier function for the brands 

def total_sales(item_name,total_items):
    if item_name=='Beef Chicharon':
        return (total_items*1299)
    elif item_name=='Nutrional Milk':
        return(total_items*1990)
    elif item_name=='Orange Beans':
        return(total_items*199)
    elif item_name=='Gummy Vitamins':
        return(total_items*1500)
    elif item_name=='Yummy Vegetables':
        return(total_items*500)
    elif item_name=='Gummy Worms':
        return(total_items*150)
    elif item_name=='Kimchi and Seaweed':
        return(total_items*799)
    
finalfinaljan['Total Sales per Item']= np.vectorize(total_sales)(finalfinaljan['item'], finalfinaljan['Total items'])

finalfinaljan
```
